```python
#function for weak antidistinguishability

def is_antidistinguishable(vectors, povm_kets, tol=1e-10):

    for i, phi in enumerate(povm_kets):
        found_orthogonal_partner = False

        for psi in vectors:
            overlap = np.vdot(phi, psi)

            if abs(overlap) < tol:
                found_orthogonal_partner = True
                break

        if not found_orthogonal_partner:
          return False

    return True
```

```python
from sympy import simplify, expand, Matrix
from sympy.physics.quantum.dagger import Dagger

def check_orthonormality(basis, tol=1e-9):
    """Returns True if the basis is orthonormal, False otherwise."""
    n = len(basis)
    for i in range(n):
        for j in range(i, n):
            # Convert numpy arrays to sympy.Matrix objects by reshaping them into column vectors
            psi_i = Matrix(basis[i].reshape(-1, 1)) # Reshape to a column vector
            phi_j = Matrix(basis[j].reshape(-1, 1)) # Reshape to a column vector

            # Inner product: <state_i | state_j>
            # Dagger(psi_i) gives the conjugate transpose of psi_i (a row vector)
            # The multiplication results in a 1x1 matrix. Access the scalar value.
            dot_product_matrix = Dagger(psi_i) * phi_j
            dot = simplify(dot_product_matrix[0]) # Access the scalar and simplify

            # Check diagonal (must be 1) and off-diagonal (must be 0)
            target = 1 if i == j else 0

            # Compare with tolerance for floating point numbers
            if abs(dot - target) > tol:
                return False
    return True

for name, basis in all_test_povms:
    # Extract only the vectors for the orthonormality check
    basis_vectors = [v for n, v in basis]
    result = "Orthonormal" if check_orthonormality(np.array(basis_vectors)) else "Not Orthonormal"
    print(f"{name}: {result}")
```
